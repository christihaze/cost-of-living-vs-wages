## National Data 

```python
df_sheltercpis = pd.read_excel('../data/shelter.xlsx', skiprows= 11)
df_medical_carecpis = pd.read_excel('../data/medicalcare.xlsx', skiprows= 11)
df_energycpis = pd.read_excel('../data/energy.xlsx', skiprows= 11)
df_food_homecpis = pd.read_excel('../data/foodhome.xlsx', skiprows= 11)
df_food_awaycpis = pd.read_excel('../data/foodaway.xlsx', skiprows= 11)
```

```python
df_sheltercpis['category']='Shelter'
df_medical_carecpis['category'] = 'Medical Care'
df_energycpis['category'] = 'Energy'
df_food_homecpis['category'] = 'Food at Home'
df_food_awaycpis['category'] = 'Food Away From Home' 
```

```python
df_all_cpis = pd.concat([df_food_homecpis, df_food_awaycpis, df_energycpis, df_sheltercpis, df_medical_carecpis], ignore_index=True)
df_all_cpis
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
df_all_cpis = pd.melt(df_all_cpis,
                   id_vars=['Year','category'],
                   value_vars=months,
                   var_name='month',
                   value_name='value')
```

```python
df_all_cpis =df_all_cpis[['Year','month','category','value']]
```

```python
df_all_cpis
```

```python
df_all_cpis= df_all_cpis.rename( columns = { 'value' : 'cpi',
                                           'Year':'year'})
df_all_cpis
```

```python
df_all_cpis['period'] = df_all_cpis['year'].apply(lambda x: 'Pre-COVID' if x <= 2019 else 'Post-COVID')
```

```python
## annual base costs
base_costs = {'Food at Home': 4015,
    'Food Away From Home': 3154,
    'Energy': 3409,
    'Shelter': 17148,
    'Medical Care': 4968}
```

```python
monthly_base_costs = {k: v/12 for k, v in base_costs.items()}
```

```python
base_cpi = df_all_cpis[df_all_cpis['year'] == 2015].set_index(['category', 'month'])['cpi'].to_dict()
```

```python
df_all_cpis['estimatedcost'] = df_all_cpis.apply(
    lambda row: monthly_base_costs[row['category']] * row['cpi'] / base_cpi[(row['category'], row['month'])],
    axis=1
)
```

```python
##average of what they paid per category per month by year
avgs=df_all_cpis.groupby(['year', 'category'], as_index=False)['estimatedcost'].mean().round(2)
avgs
```

```python
## gives me the sum of each category per year
grouped = df_all_cpis.groupby(['year', 'category'], as_index=False)['estimatedcost'].sum().round(2)
grouped
```

```python
pivot_grouped = grouped.pivot(index='year', columns='category', values='estimatedcost')
```

```python
pivot_grouped.plot(kind='line', marker='o')
plt.title('Changes Over Years by Category')
plt.xlabel('Year')
plt.ylabel('Estimated Cost')
plt.grid(True)
plt.legend(title='Category')
plt.show()
```

```python
pivot_avgs = avgs.pivot(index='year', columns='category', values='estimatedcost')
```

```python
pivot_avgs.plot(kind='line', marker='o')
plt.title('Changes by Monthly Average')
plt.xlabel('Year')
plt.ylabel('Estimated Cost')
plt.grid(True)
plt.legend(title='Category')
plt.show()
```

```python
df_all_cpis
```

```python
pre_v_post = df_all_cpis.groupby(['category', 'period'])['estimatedcost'].mean().reset_index().round(2)
pre_v_post
```

```python
df_all_cpis ['estimatedcost'] = df_all_cpis['estimatedcost'].round(2)
```

```python
df_all_cpis['region']= 'National'
df_all_cpis['month']= df_all_cpis['month'].str.lower()
df_all_cpis
```
